### Créer un Mongo Client afin d'exécuter nos requêtes sur MONGODB Server

```python
client = pymongo.MongoClient('mongodb://127.0.0.1:27017/')
```

###  Afficher les différentes  databases

```python
client.list_database_names()
```

### Créer une "database variable" pour travailler avec la BDD "dblp"

```python
db = client["dblp"]
```

### Récupérer les collections de la BDD "dblp"

```python
db.list_collection_names()
```

### Donner le titre des publications dont le type est « Book »

```python
cursor = db["publis"].find({'type':'Book'}).limit(10)
for pub in cursor:
    print(pub)
```

### Donner le titre des publications ayant «publisher » et « editor » renseignés



```python
cursor = db["publis"].find({'publisher':{'$exists':True},'editor':{'$exists':True}}).limit(1)
for pub in cursor:
    print(pub)
```

### 	Donner le titre des publications qui sont publiées entre 2005 et 2007.

```python
#method1:
cursor = db["publis"].find({'$and':[{'year':{'$gte':2005}},{'year':{'$lte':2007}}]  },{'year':1}).limit(30)
for pub in cursor:
    print(pub)
```

```python
#method2
cursor = db["publis"].find({'year':{'$gte':2005,'$lte':2007}  },{'year':1}).limit(30)
for pub in cursor:
    print(pub)
```

### Donner le nombre des publications dont le titre contient « Ontology » 


```python
cursor = db["publis"].find({"title":{"$regex":"Ontology","$options":"i"} },{"title":1}).limit(30)
for pub in cursor:
    print(pub)
```

### Donner le titre des publications publiées par l’auteur «Mimoun Malki »



```python
cursor = db["publis"].find({"authors":"Mimoun Malki" },{"title":1,"authors":1,"_id":0}).limit(30)
for pub in cursor:
    print(pub)
```

### Donner le titre des publications dont  «Mimoun Malki » est premier auteur

```python
cursor = db["publis"].find({"authors.0":"Mimoun Malki" },{"title":1,"authors":1,"_id":0}).limit(30)
for pub in cursor:
    print(pub)
```

### Liste distincte des auteurs



```python
db["publis"].distinct("authors")
```

### Retourner le titre et l’année des publications  de type article, et trier sur l'année dans le l’ordre décroissant  


```python
cursor = db["publis"].find({"type":"Article" },{"title":1,"year":1,"_id":0}).sort([("year", pymongo.DESCENDING)]).limit(30)
for pub in cursor:
    print(pub)
```

### Donner le nombre publications par type



```python
cursor = db["publis"].aggregate([{"$group":{"_id":"$type","nb_publication":{"$sum":1}}}])
for pub in cursor:
    print(pub)
```

### Pour chaque auteur le nombre des publications. Trier par ordre décroissant




```python
cursor = db["publis"].aggregate([
    {"$unwind":"$authors"},
    {"$group":{"_id":"$authors","nb_publication":{"$sum":1}}},
    {"$sort":{"nb_publication":-1}}
])
for pub in cursor:
    print(pub)
```

## le nombre de pages moyen par type

```python
cursor = db["publis"].aggregate([
    {"$group":{"_id":"$type","avg_nb_pages":{"$avg":{"$subtract":["$pages.end","$pages.start"]}}}}
])
for pub in cursor:
    print(pub)
```
